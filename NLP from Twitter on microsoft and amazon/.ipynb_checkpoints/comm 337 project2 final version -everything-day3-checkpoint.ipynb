```python
%matplotlib inline
!pip install textblob
!pip install wordcloud
from textblob import TextBlob

import json
import string
import nltk
import os

def rm_dp(text):
    p = string.punctuation.replace('#','').replace('@','')
    d = string.digits
    table_p = str.maketrans(p, len(p) * " ")
    table_d = str.maketrans(d, len(d) * " ")
    
    text = text.translate(table_d).translate(table_p)
    newtext=""
    for t in text:
        if t in string.printable:
            newtext=newtext+t
    return newtext
def rm_stopword(lst):#remove stopwords
    stopwords = nltk.corpus.stopwords.words('english')
    ct=[]
    for sentences in lst:
        newlst = []
        for w in sentences.lower().split():
            if w not in stopwords and len(w) > 1:
                newlst.append(w)
        clean_sentences=""
        for w in newlst:
            clean_sentences+=w+" "
        ct.append(clean_sentences)
    return ct

def rm_meaningless(lst):#remove meaningless words
    exwords=["https","co","rt","http",'amp','de','theyr']
    ct=[]
    for sentences in lst:
        newlst = []
        for w in sentences.lower().split():
            if w not in exwords:
                newlst.append(w)
        clean_sentences=""
        for w in newlst:
            clean_sentences+=w+" "
        ct.append(clean_sentences)
    return ct

def clean(lst):
    cleantext=[]
    for item in lst:
        cleantext.append(rm_dp(item['text']))
    cleantext=rm_stopword(cleantext)
    cleantext=rm_meaningless(cleantext)
    return cleantext

def notclean(lst):
    notcleantext=[]
    for item in lst:
        notcleantext.append(rm_dp(item['text']))
    return notcleantext

def loading(keyword):#loading JSON files
    all_tweets=[]
    PATH = './JSON2/day3'
    for fname in sorted(os.listdir(PATH)):
        if fname.endswith(keyword+'.json'):
            with open('{}/{}'.format(PATH, fname)) as infile:
                tweets = json.load(infile)
                all_tweets.extend(tweets)
    return all_tweets

#tranlate to en
def trans_en(lst):
    tweet=[]
    for item in all_tweets:
        if item['lang'] != 'en':
            tb = TextBlob(item['text'])
            translated=tb.translate(to='en')
        tweet.append(tranlated)
    return tweet

#Stemming 词根转换
from nltk.stem.snowball import SnowballStemmer
ss = SnowballStemmer("english") 
```

```python
all_tweets=loading('microsoft')
#if you want to translate all tweets to english
#all_tweets=trans_en(all_tweets)
cleantext=clean(all_tweets)
notcleantext=notclean(all_tweets)
```

```python
##摘选出来两组词，第一组words是包括@ 和#的，words_without_hash则是没有包括任何符号的

content=""
for sentences in cleantext:
    content+=sentences
p = string.punctuation
table_p = str.maketrans(p, len(p) * " ")
words=content.split()
content = content.translate(table_p)
words_without_hash=content.split()

words_stemmed=[]
for w in words_without_hash:
    words_stemmed.append(ss.stem(w))

#Q1 Most frequent words without stop words
freq = nltk.FreqDist(words_stemmed)
print('Top 10 most frequent words without stop words')
freq.plot(10)

content2=""
for sentences2 in notcleantext:
    content2+=sentences2
p = string.punctuation
table_p = str.maketrans(p, len(p) * " ")
words2=content2.split()
content2 = content2.translate(table_p)
words_without_hash2=content2.split()

words_stemmed2=[]
for w in words_without_hash2:
    words_stemmed2.append(ss.stem(w))
    
#Q1 Most frequent words with stop words
freq = nltk.FreqDist(words_stemmed2)
print('Top 10 most frequent words with stop words')
freq.plot(10)

#Q2 hashtage
q2=[ t for t in words if t.startswith('#') ]
freq = nltk.FreqDist(q2)
print('Top 10 most popluar hashtage')
freq.plot(10)

#Q3 mention@
q3=[ t for t in words if t.startswith('@') ] 
freq = nltk.FreqDist(q3)
print('Top 10 most popluar metion@')
freq.plot(10) 

#Q4 most frequently tweeting person
user_lst=[]
for item  in all_tweets:
    user_lst.append(item['user']['name'])
freq = nltk.FreqDist(user_lst)
print('{} {}'.format('The most frequently tweeting person is',freq.max()))
```

```python
#Q5 most influential tweet?
l1=[]
for item in all_tweets:
    if 'retweeted_status' in item:
        l1.append(item['retweeted_status']['retweet_count'])
    else:
        l1.append(item['retweet_count'])
    
l2=[]
for item1 in all_tweets:
    l2.append(item1['reply_count'])

l3=[]
for item2 in all_tweets:
    l3.append(item2['favorite_count'])
    
l4=[]
for item3 in all_tweets:
    l4.append(item3['quote_count'])

score=[]
for i in range(len(l1)):
    score.append(l1[i]+l2[i]+l3[i]+l4[i])
    
q5=all_tweets[score.index(max(score))]['text']
print("{} {}".format('the most influential tweet is this:\n',q5))
```

```python
#PartC
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from textblob import TextBlob
#将all tweets转换成一句一句话
sentence=[]

for i in all_tweets:
    sentence.append(i['text'])

#没有用到cleantext来做这个分析，因为stopwords其实不影响这个Subjectivity判断的

#画Subjectivity & Polarity的图
sub_list = []
pol_list = []
for s in sentence:
    tb = TextBlob(s)
    sub_list.append(tb.sentiment.subjectivity)
    pol_list.append(tb.sentiment.polarity)

import matplotlib.pyplot as plt

plt.hist(sub_list, bins=10) #, normed=1, alpha=0.75)
plt.xlabel('subjectivity score')
plt.ylabel('sentence count')
plt.grid(True)
plt.show()


plt.hist(pol_list, bins=10) #, normed=1, alpha=0.75)
plt.xlabel('polarity score')
plt.ylabel('sentence count')
plt.grid(True)
plt.show()

avg_subjectivity=sum(sub_list)/len(sub_list)
avg_polarity=sum(pol_list)/len(pol_list)
print("{} {} {} {}".format('The average subjectivity is:',avg_subjectivity,'The average polarity is:',avg_polarity))
```
